```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import csv
```

```python
df = pd.read_csv("Data1.csv", index_col=0, parse_dates=True)
df.head()
```

```python
cols_plot = ['theta_1','theta_2','theta_3','theta_4','theta_5','theta_6']
axes = df[cols_plot].plot(marker='.', alpha=0.5, subplots=True, figsize=(17,20))
```

```python
df_melted = df.melt()
```

```python
sns.set(style="darkgrid")
fg = sns.FacetGrid(df_melted, col="variable", sharex=False, sharey=False);
fg.map(plt.hist, "value", color="steelblue");
```

```python
fg = sns.FacetGrid(df_melted, col="variable", sharex=False, sharey=False);
fg.map(sns.kdeplot, 'value')
```

```python
df_2018 = df.loc['2018-01-01':'2018-12-31']
```

```python
axes = df_2018[cols_plot].plot(marker='.', alpha=0.5, subplots=True, figsize=(17,20))
```

```python
df_2018_melted = df_2018.melt()
```

```python
fg = sns.FacetGrid(df_2018_melted, col="variable", sharex=False, sharey=False);
fg.map(plt.hist, "value", color="steelblue");
```
