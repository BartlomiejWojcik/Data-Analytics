```python
import pandas as pd 
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import logit, expit
```

 # Task 1 - Normal distribution 
 

### The larger number of bins results in a better match between the data received and the expected normal distribution. Changing the mean shift entire distribution left or right, according to mean value. Changing the variance tightens or spreads out the width of the distribution along the X-axis. Larger standard deviations produce distributions that are more spread out.

```python
np.random.seed(42)
```

```python
def normal_distribution_histograms(mu,sigma):
    samples = np.random.normal(mu,sigma,1000)
    plt.subplot(3, 2, 1)
    plt.grid(True)
    plt.hist(samples, 10, density=True)
    plt.title('cumulative=False')
    plt.subplot(3, 2, 2)
    plt.grid(True)
    plt.hist(samples, 10, density=True, cumulative=True)
    plt.title('cumulative=True')
    plt.subplot(3, 2, 3)
    plt.grid(True)
    plt.hist(samples, 25, density=True)
    plt.subplot(3, 2, 4)
    plt.grid(True)
    plt.hist(samples, 25, density=True, cumulative=True)
    plt.subplot(3, 2, 5)
    plt.grid(True)
    plt.hist(samples, 100, density=True)
    plt.subplot(3, 2, 6)
    plt.grid(True)
    plt.hist(samples, 100, density=True, cumulative=True)
    plt.grid(True)
```

```python
normal_distribution_histograms(0,0.1)
```

```python
normal_distribution_histograms(10,0.1)
```

```python
normal_distribution_histograms(0,10)
```

 # Task 2 - Inverse logit
 

### In case of inverse logit, the shape of the decomposition changes significantly, as the variance increases. Most of the generated data is located on the edges. More to the left for mean < 0 and more on the right for mean > 0.

```python
def normal_distribution_inverse_logit_histograms(mu,sigma):
    samples = np.random.normal(mu,sigma,1000)
    samples = expit(samples)
    plt.subplot(3, 2, 1)
    plt.grid(True)
    plt.hist(samples, 10, density=True)
    plt.title('cumulative=False')
    plt.subplot(3, 2, 2)
    plt.grid(True)
    plt.hist(samples, 10, density=True, cumulative=True)
    plt.title('cumulative=True')
    plt.subplot(3, 2, 3)
    plt.grid(True)
    plt.hist(samples, 25, density=True)
    plt.subplot(3, 2, 4)
    plt.grid(True)
    plt.hist(samples, 25, density=True, cumulative=True)
    plt.subplot(3, 2, 5)
    plt.grid(True)
    plt.hist(samples, 100, density=True)
    plt.subplot(3, 2, 6)
    plt.grid(True)
    plt.hist(samples, 100, density=True, cumulative=True)
    plt.grid(True)
```

```python
normal_distribution_inverse_logit_histograms(0,0.1)
```

```python
normal_distribution_inverse_logit_histograms(0,3)
```

```python
normal_distribution_inverse_logit_histograms(-4,5)
```

```python
normal_distribution_inverse_logit_histograms(3,5)
```

# Task 3 - Poisson distribution

###  In this distribution, the mean equals the variance and is determined by the lambda parameter. Changing the variance tightens distribution, according to given locations of bins.

```python
def poisson_distribution_inverse_logit_histograms(lam):
    samples = np.random.poisson(lam,1000)
    plt.subplot(3, 2, 1)
    plt.grid(True)
    plt.hist(samples,np.linspace(0, lam*3, lam), density=True)
    plt.title('cumulative=False')
    plt.subplot(3, 2, 2)
    plt.grid(True)
    plt.hist(samples, np.linspace(0, lam*3, lam), density=True, cumulative=True)
    plt.title('cumulative=True')
    plt.subplot(3, 2, 3)
    plt.grid(True)
    plt.hist(samples, np.linspace(0, lam*2, lam*2), density=True)
    plt.subplot(3, 2, 4)
    plt.grid(True)
    plt.hist(samples, np.linspace(0, lam*2, lam*2), density=True, cumulative=True)
    plt.subplot(3, 2, 5)
    plt.grid(True)
    plt.hist(samples, np.linspace(0, lam*3, lam*3), density=True)
    plt.subplot(3, 2, 6)
    plt.grid(True)
    plt.hist(samples, np.linspace(0, lam*3, lam*3), density=True, cumulative=True)
    plt.grid(True)
```

```python
poisson_distribution_inverse_logit_histograms(5)
```

```python
poisson_distribution_inverse_logit_histograms(20)
```

```python
poisson_distribution_inverse_logit_histograms(50)
```

# Task 4 - Beta distribution

### The distribution is symmetrical for alpha = beta with mean = 1/2. In this case when alpha and beta are more than 1 distribution is like normal distribution but for less than 1 is like Inverse logit. As the proportions of the parameters are going to infinite or infinitely small values, the distribution becomes more skewness (right or left). Its behaviour also depends on parametrization method.

```python
def beta_distribution_histograms(a,b):
    samples = np.random.beta(a,b,1000)
    plt.subplot(3, 2, 1)
    plt.grid(True)
    plt.hist(samples,10, density=True)
    plt.title('cumulative=False')
    plt.subplot(3, 2, 2)
    plt.grid(True)
    plt.hist(samples, 10, density=True, cumulative=True)
    plt.title('cumulative=True')
    plt.subplot(3, 2, 3)
    plt.grid(True)
    plt.hist(samples, 25, density=True)
    plt.subplot(3, 2, 4)
    plt.grid(True)
    plt.hist(samples, 25, density=True, cumulative=True)
    plt.subplot(3, 2, 5)
    plt.grid(True)
    plt.hist(samples, 100, density=True)
    plt.subplot(3, 2, 6)
    plt.grid(True)
    plt.hist(samples, 100, density=True, cumulative=True)
    plt.grid(True)
```

```python
def beta_location_distribution_histograms(mi,phi):
    alpha = (1-phi)*mi/phi
    beta = (1-mi)*(1-phi)/phi
    samples = np.random.beta(alpha,beta,1000)
    plt.subplot(3, 2, 1)
    plt.grid(True)
    plt.hist(samples,10, density=True)
    plt.title('cumulative=False')
    plt.subplot(3, 2, 2)
    plt.grid(True)
    plt.hist(samples, 10, density=True, cumulative=True)
    plt.title('cumulative=True')
    plt.subplot(3, 2, 3)
    plt.grid(True)
    plt.hist(samples, 25, density=True)
    plt.subplot(3, 2, 4)
    plt.grid(True)
    plt.hist(samples, 25, density=True, cumulative=True)
    plt.subplot(3, 2, 5)
    plt.grid(True)
    plt.hist(samples, 100, density=True)
    plt.subplot(3, 2, 6)
    plt.grid(True)
    plt.hist(samples, 100, density=True, cumulative=True)
    plt.grid(True)
```

# Task 5 - Log-normal distribution 

### logarithm of these distributions is normally distributed. As the variance increases, the distribution moves towards zero and becomes wider. 

```python
def lognormal_distribution_histograms(mean,sigma):
    samples = np.random.lognormal(mean,sigma,1000)
    plt.subplot(3, 2, 1)
    plt.grid(True)
    plt.hist(samples,10, density=True)
    plt.title('cumulative=False')
    plt.subplot(3, 2, 2)
    plt.grid(True)
    plt.hist(samples, 10, density=True, cumulative=True)
    plt.title('cumulative=True')
    plt.subplot(3, 2, 3)
    plt.grid(True)
    plt.hist(samples, 25, density=True)
    plt.subplot(3, 2, 4)
    plt.grid(True)
    plt.hist(samples, 25, density=True, cumulative=True)
    plt.subplot(3, 2, 5)
    plt.grid(True)
    plt.hist(samples, 100, density=True)
    plt.subplot(3, 2, 6)
    plt.grid(True)
    plt.hist(samples, 100, density=True, cumulative=True)
    plt.grid(True)
```
