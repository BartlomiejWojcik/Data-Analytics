```python
import pandas as pd 
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import logit, expit
```

```python
np.random.seed(42)
```

```python
def normal_distribution_histograms(mu,sigma):
    samples = np.random.normal(mu,sigma,1000)
    plt.subplot(3, 2, 1)
    plt.grid(True)
    plt.hist(samples, 10, density=True)
    plt.title('cumulative=False')
    plt.subplot(3, 2, 2)
    plt.grid(True)
    plt.hist(samples, 10, density=True, cumulative=True)
    plt.title('cumulative=True')
    plt.subplot(3, 2, 3)
    plt.grid(True)
    plt.hist(samples, 25, density=True)
    plt.subplot(3, 2, 4)
    plt.grid(True)
    plt.hist(samples, 25, density=True, cumulative=True)
    plt.subplot(3, 2, 5)
    plt.grid(True)
    plt.hist(samples, 100, density=True)
    plt.subplot(3, 2, 6)
    plt.grid(True)
    plt.hist(samples, 100, density=True, cumulative=True)
    plt.grid(True)
```

```python
normal_distribution_histograms(0,0.1)
```

```python
normal_distribution_histograms(10,0.1)
```

```python
normal_distribution_histograms(0,10)
```

```python
def normal_distribution_inverse_logit_histograms(mu,sigma):
    samples = np.random.normal(mu,sigma,1000)
    samples = expit(samples)
    plt.subplot(3, 2, 1)
    plt.grid(True)
    plt.hist(samples, 10, density=True)
    plt.title('cumulative=False')
    plt.subplot(3, 2, 2)
    plt.grid(True)
    plt.hist(samples, 10, density=True, cumulative=True)
    plt.title('cumulative=True')
    plt.subplot(3, 2, 3)
    plt.grid(True)
    plt.hist(samples, 25, density=True)
    plt.subplot(3, 2, 4)
    plt.grid(True)
    plt.hist(samples, 25, density=True, cumulative=True)
    plt.subplot(3, 2, 5)
    plt.grid(True)
    plt.hist(samples, 100, density=True)
    plt.subplot(3, 2, 6)
    plt.grid(True)
    plt.hist(samples, 100, density=True, cumulative=True)
    plt.grid(True)
```

```python
normal_distribution_inverse_logit_histograms(0,0.1)
```

```python
normal_distribution_inverse_logit_histograms(10,0.1)
```

```python
normal_distribution_inverse_logit_histograms(0,1)
```

```python
normal_distribution_inverse_logit_histograms(0,3)
```

```python
def poisson_distribution_inverse_logit_histograms(lam):
    samples = np.random.poisson(lam,1000)
    plt.subplot(3, 2, 1)
    plt.grid(True)
    plt.hist(samples,np.linspace(0, lam*3, lam), density=True)
    plt.title('cumulative=False')
    plt.subplot(3, 2, 2)
    plt.grid(True)
    plt.hist(samples, np.linspace(0, lam*3, lam), density=True, cumulative=True)
    plt.title('cumulative=True')
    plt.subplot(3, 2, 3)
    plt.grid(True)
    plt.hist(samples, np.linspace(0, lam*2, lam*2), density=True)
    plt.subplot(3, 2, 4)
    plt.grid(True)
    plt.hist(samples, np.linspace(0, lam*2, lam*2), density=True, cumulative=True)
    plt.subplot(3, 2, 5)
    plt.grid(True)
    plt.hist(samples, np.linspace(0, lam*3, lam*3), density=True)
    plt.subplot(3, 2, 6)
    plt.grid(True)
    plt.hist(samples, np.linspace(0, lam*3, lam*3), density=True, cumulative=True)
    plt.grid(True)
```

```python
def beta_location_distribution_histograms(mi,phi):
    alpha = (1-phi)*mi/phi
    beta = (1-mi)*(1-phi)/phi
    samples = np.random.beta(alpha,beta,1000)
    plt.subplot(3, 2, 1)
    plt.grid(True)
    plt.hist(samples,10, density=True)
    plt.title('cumulative=False')
    plt.subplot(3, 2, 2)
    plt.grid(True)
    plt.hist(samples, 10, density=True, cumulative=True)
    plt.title('cumulative=True')
    plt.subplot(3, 2, 3)
    plt.grid(True)
    plt.hist(samples, 25, density=True)
    plt.subplot(3, 2, 4)
    plt.grid(True)
    plt.hist(samples, 25, density=True, cumulative=True)
    plt.subplot(3, 2, 5)
    plt.grid(True)
    plt.hist(samples, 100, density=True)
    plt.subplot(3, 2, 6)
    plt.grid(True)
    plt.hist(samples, 100, density=True, cumulative=True)
    plt.grid(True)
```

```python
def beta_population_distribution_histograms(n):
    alpha = 0.485*n
    beta = n-alpha
    samples = np.random.beta(438+alpha,544+beta,1000)
    plt.subplot(3, 2, 1)
    plt.grid(True)
    plt.hist(samples,10, density=True)
    plt.title('cumulative=False')
    plt.subplot(3, 2, 2)
    plt.grid(True)
    plt.hist(samples, 10, density=True, cumulative=True)
    plt.title('cumulative=True')
    plt.subplot(3, 2, 3)
    plt.grid(True)
    plt.hist(samples, 25, density=True)
    plt.subplot(3, 2, 4)
    plt.grid(True)
    plt.hist(samples, 25, density=True, cumulative=True)
    plt.subplot(3, 2, 5)
    plt.grid(True)
    plt.hist(samples, 100, density=True)
    plt.subplot(3, 2, 6)
    plt.grid(True)
    plt.hist(samples, 100, density=True, cumulative=True)
    plt.grid(True)
```
