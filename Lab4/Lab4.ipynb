# Task 1 
## The rate is constant for all years and not depending on anything.

Starting from very simple model. Assuming, that passanger deaths number $y_i$ has a Poisson distribution 
$$y_i\sim\mathrm{Poisson}(\lambda)$$
with a rate $\lambda$ independent on year or miles flown.

It can be assumed that there is a very small chance that the number of passengers deaths in one year is more than 15000, so that chance is less than 1%.
For poisson distribution we have mean of $\lambda$ and standard deviation of $\sqrt{\lambda}$. Approximately in order to have no more than 1% probability $\lambda$ should fulfill
$$\lambda+3\sqrt{\lambda}\approx15000$$
We need to assign the prior that would have probability of smaller $\lambda$ equal 99%.

```python
root_of_lam=np.polynomial.polynomial.polyroots([-15000.,3.,1.])
lam_ub=np.round(root_of_lam[root_of_lam>0]**2)
print(lam_ub)
```

This means that the lambda parameter should be between 0 and 14637.

The next step is to find the right $\sigma$ for a HalfNormal distribution

```python
with open('prior_model.stan', 'r') as file:
    print(file.read())
```

```python
tuning=stan_utility.compile_model('prior_model.stan')
```

```python
data=dict(y_guess=np.array([np.log(10000)]),theta=np.array(lam_ub))
tuned1 = tuning.sampling(data=data,
                          seed=1228022,
                          algorithm="Fixed_param",
                          iter=1, 
                          warmup=0,
                          chains=1)
sigma = np.round(tuned1.extract()['sigma'][0])
print(sigma)
```

```python
fig, ax2 = plt.subplots(1, 1,figsize=(7, 4))
x2=np.linspace(0,3*sigma,1000)
x4=np.linspace(0,lam_ub[0],1000)
ax2.plot(x2,2*stats.norm.pdf(x2,scale=sigma),color=dark,linewidth=2)
ax2.fill_between(x4,2*stats.norm.pdf(x4,scale=sigma),0,color=dark)
ax2.set_yticks([])
ax2.set_xticks([0,lam_ub[0]])
ax2.set_title(r'$\lambda$')
plt.show()
```

# Task 2
## The intensity is constant, but rate is a product of intensity and miles flown.

Assuming that number of accidents should be related to number of miles flown, we can again using Poisson Distribution using intensity $\theta$ and exposure $n$. $$y_i\sim\mathrm{Poisson}(\theta n)$$
With $n$ being a number miles flown (in 100 mil).
Still assuming that 15000 deaths in one year would be very improbable. In order to compute the bound we use $\lambda=\theta\cdot\bar{n}$, with $\bar{n}$ being mean of miles flown. This gives us condition

$$\theta\cdot\bar{n}+3\sqrt{\theta\cdot\bar{n}}\approx15000$$

We need to assign the prior for $\theta$ that would have probability of smaller $\lambda$ equal 99%.

```python
mean_miles=np.mean(Airline_data['Miles flown [100 mln miles]'])
root_of_theta=np.polynomial.polynomial.polyroots([-15000/mean_miles,3./np.sqrt(mean_miles),1.])
theta_ub=(root_of_theta[root_of_lam>0]**2)
print('theta upper bound','{:4.3f}'.format(theta_ub[0]))
```

```python
data=dict(y_guess=np.array([np.log(0.1)]),theta=np.array(theta_ub))
tuned2 = tuning.sampling(data=data,
                          seed=1228022,
                          algorithm="Fixed_param",
                          iter=1, 
                          warmup=0,
                          chains=1)
sigma = (tuned2.extract()['sigma'][0])
print('sigma =','{:4.3f}'.format(sigma))
```

```python
fig, ax2 = plt.subplots(1, 1,figsize=(7, 4))
x2=np.linspace(0,3*sigma,1000)
x4=np.linspace(0,theta_ub[0],1000)
ax2.plot(x2,2*stats.norm.pdf(x2,scale=sigma),color=dark,linewidth=2)
ax2.fill_between(x4,2*stats.norm.pdf(x4,scale=sigma),0,color=dark)
ax2.set_yticks([])
ax2.set_xticks([0,theta_ub[0]])
ax2.set_title(r'$\theta$')
plt.show()
```
